```python
from __future__ import print_function, division
%matplotlib inline
import os
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import geopandas as gpd

DEVELOPE = True
```

```python
#taxi
df = pd.read_csv('data/clean_taxi_data.csv', index_col=0)
print ("Shape of data\n{}".format(df.shape))
df.head()
```

```python
#FHV
df_fhv = pd.read_csv('data/clean_FHV_data.csv', index_col=0)
print ("Shape of data\n{}".format(df_fhv.shape))
df_fhv.head()
```

```python
#weather
df_wea = pd.read_csv('data/clean_weather_data.csv', index_col=0)
print ("Shape of data\n{}".format(df_wea.shape))
df_wea.head()
```

## Merge

```python
df = df.merge(df_fhv, on='day_and_hour', how='left')
df = df.merge(df_wea, on='dayofyear', how='left')
```

```python
#not meaningful in this small time frame
df.drop(['glaze_rime', 'mist', 'drizzle', 'ice_fog', 'year', 'month'], axis=1, inplace=True)
```
