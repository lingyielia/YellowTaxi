```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%pylab inline
```

```python
sorted_features = [('FHV_count', 0.28776649763404427),
 ('longitude', 0.26250518029388664),
 ('latitude', 0.18562738884675992),
 ('wind_speed_average', 0.042727152730180171),
 ('temperature_max', 0.035902088940227041),
 ('temperature_min', 0.034766226162321397),
 ('fastest_2min_windspeed', 0.029261392096689966),
 ('dayofyear', 0.028177373083360671),
 ('DAILYSunset', 0.026283298698263867),
 ('DAILYSunrise', 0.025142296680793711),
 ('precipitation', 0.018001517096841519),
 ('snow_depth', 0.0082554751314433128),
 ('smoke_haze', 0.0052297155239492124),
 ('fog', 0.004097123119361795),
 ('month', 0.0030597961476927643),
 ('snowfall', 0.0030511773429814171),
 ('ice_pellets', 0.00014630047120243279)]
```

```python
# Most important features
plt.figure(figsize=(20,10))
sorted_features.sort(key=lambda x: x[1], reverse=True) 

features = list(zip(*sorted_features))[0]
score = list(zip(*sorted_features))[1]
x_pos = np.arange(len(features)) 

  
plt.bar(x_pos, score,align='center')
plt.xticks(x_pos, features, rotation=70, size=20) 
plt.ylabel('Feature Importance', size=20)
plt.title('Feature Importance by Type', size=24)
plt.show()
```
