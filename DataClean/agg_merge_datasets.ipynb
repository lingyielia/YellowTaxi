```python
from __future__ import print_function, division
%matplotlib inline
import os
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import geopandas as gpd

DEVELOPE = True
```

```python
#taxi
df = pd.read_csv('data/agg_clean_taxi_201501_06.csv', index_col=0)
print ("Shape of data\n{}".format(df.shape))
df.head()
```

```python
#FHV
df_fhv = pd.read_csv('data/agg_clean_FHV_201501_06.csv', index_col=0)
print ("Shape of data\n{}".format(df_fhv.shape))
df_fhv.head()
```

```python
#weather
df_wea = pd.read_csv('data/clean_weather_data_2015.csv', index_col=0)
print ("Shape of data\n{}".format(df_wea.shape))
df_wea.head()
```

## Merge

```python
print ("Shape of data{}".format(df.shape))
df = df.merge(df_fhv, on='day_and_hour', how='left')
df.drop_duplicates(inplace=True)
print ("Shape of data{}".format(df.shape))
df = df.merge(df_wea, on='dayofyear', how='left')
df.drop_duplicates(inplace=True)
print ("Shape of data{}".format(df.shape))
```
